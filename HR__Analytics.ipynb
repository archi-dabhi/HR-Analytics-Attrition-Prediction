## **Handling Outlier**
---
## 5.4 Monthly Income vs Attrition (Boxplot with Outlier Handling)

```python

def cap_outliers(x):
    Q1, Q3 = x.quantile([0.25,0.75])
    IQR = Q3 - Q1
    low, up = Q1 - 1.5*IQR, Q3 + 1.5*IQR
    return np.where(x > up, up, np.where(x < low, low, x))

data['MonthlyIncome_cap'] = data.groupby('Attrition')['MonthlyIncome'].transform(cap_outliers)

# Boxplot
sns.boxplot(x='Attrition', y='MonthlyIncome_cap', data=data, palette="viridis")
plt.title("Monthly Income by Attrition (Group-wise Outliers Capped)")
plt.show()
```

## Observation :

* Employees who left (Attrition = Yes) still generally have lower monthly incomes compared to those who stayed.  
* High-income employees (above ~10,000) continue to show very low attrition rates, confirming salary as a key factor in retention.  
* Group-wise outlier capping has **reduced extreme values**, making the distribution within each Attrition group more compact.  
* The median and spread of MonthlyIncome are now clearer, allowing easier comparison between Attrition groups without distortion from outliers.

# **7. Data Encoding & Scaling**
---

* Encoding: Converts categorical variables into numeric form (e.g., Attrition Yes/No → 1/0 using Label Encoding, and Department, JobRole → dummy variables using One-Hot Encoding).

* Scaling: Standardizes numerical features (mean = 0, standard deviation = 1) using StandardScaler to ensure all variables are on the same scale, improving model performance.

```python
# -----------------------------
# 7: Data Encoding & Scaling
# -----------------------------

import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler


le = LabelEncoder()
data['Attrition'] = le.fit_transform(data['Attrition'])

categorical_cols = data.select_dtypes(include='object').columns
df = pd.get_dummies(data, columns=categorical_cols, drop_first=True)


# Separate features & target
X = df.drop('Attrition', axis=1)
y = df['Attrition']


# Scale numeric features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Completed : Encoding & Scaling done")
```

# **8. Train/Test Split & Baseline Model**
---

* Train/Test Split: The dataset is divided into training (to train the model)
and testing sets (to evaluate performance). Typically, 70–80% is used for training and 20–30% for testing.

* Baseline Model: A simple model (like Logistic Regression) is trained first to set a performance benchmark. Later, more complex models can be compared against this baseline.



```python
# -----------------------------
# 8: Train/Test Split & Baseline Model
# -----------------------------

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

print("Overall class distribution:\n", y.value_counts())


# Split data (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)


model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Heatmap
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',xticklabels=['No', 'Yes'],yticklabels=['No', 'Yes'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix Heatmap")
plt.show()


print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1-score:", f1_score(y_test, y_pred))

```

# **9. Additional Models & Comparison**
---

Train different models (e.g., Logistic Regression, Decision Tree, Random Forest, SVM) and compare their performance using metrics (accuracy, precision, recall, F1). This helps select the best model for the problem.

- **Decision Tree** → A simple model that splits data into rules, making it easy to understand and interpret.

- **Random Forest** → An ensemble of many trees that reduces overfitting and improves accuracy.

- **Cross-Validation** → 5-fold CV provides reliable and unbiased performance estimation.

- **ROC & AUC** → Random Forest shows better discrimination ability with higher AUC and accuracy



```python
# -----------------------------
# 9: Additional Models & Comparison
# -----------------------------

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# --- Decision Tree ---

dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)

print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt))
print("Decision Tree Report:\n", classification_report(y_test, y_pred_dt))

# --- Random Forest ---

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Random Forest Report:\n", classification_report(y_test, y_pred_rf))

# --- Cross-validation ---

cv_dt = cross_val_score(dt, X_scaled, y, cv=5).mean()
cv_rf = cross_val_score(rf, X_scaled, y, cv=5).mean()
print("CV Decision Tree:", cv_dt)
print("CV Random Forest:", cv_rf)

# --- ROC Curve ---

y_prob_dt = dt.predict_proba(X_test)[:,1]
y_prob_rf = rf.predict_proba(X_test)[:,1]

fpr_dt, tpr_dt, _ = roc_curve(y_test, y_prob_dt)
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)

plt.plot(fpr_dt, tpr_dt, label="Decision Tree (AUC = %.2f)" % roc_auc_score(y_test, y_prob_dt))
plt.plot(fpr_rf, tpr_rf, label="Random Forest (AUC = %.2f)" % roc_auc_score(y_test, y_prob_rf))
plt.plot([0,1],[0,1],'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

# **10. Model Selection & Tuning**  
---

The process of finding the best hyperparameters (settings) for a model using methods like GridSearchCV with cross-validation to improve performance and reliability.

```python
# -----------------------------
# 10: Model Selection & Tuning
# -----------------------------

from sklearn.model_selection import GridSearchCV

# Define grid
param_grid = {
    "n_estimators": [100, 200, 300],
    "max_depth": [None, 5, 10, 20],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}

grid_search = GridSearchCV(RandomForestClassifier(random_state=42),
                           param_grid,
                           cv=5,
                           scoring='accuracy',
                           n_jobs=-1,
                           verbose=2)

grid_search.fit(X_train, y_train)

print("Best Parameters:", grid_search.best_params_)
print("Best CV Score:", grid_search.best_score_)

# Evaluate best model
best_rf = grid_search.best_estimator_
y_pred_best = best_rf.predict(X_test)

print("Tuned Random Forest Accuracy:", accuracy_score(y_test, y_pred_best))
print("Tuned Random Forest Report:\n", classification_report(y_test, y_pred_best))
```

# **11. Explainability with SHAP**

SHAP shows **how each feature impacts the model’s predictions**, using **dot and bar plots** to display feature importance and effect.


```python
# -----------------------------
# 11: Explainability with SHAP
# -----------------------------


from sklearn.ensemble import RandomForestClassifier
import shap


best_rf = RandomForestClassifier(random_state=42)
best_rf.fit(X_train, y_train)

# SHAP
explainer = shap.TreeExplainer(best_rf)
shap_values = explainer.shap_values(X_test)

# For binary classification
if isinstance(shap_values, list):
    shap_values = shap_values[1]

shap.summary_plot(shap_values, X_test, feature_names=X.columns)
shap.summary_plot(shap_values, X_test, feature_names=X.columns, plot_type="bar")
```
