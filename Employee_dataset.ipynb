```python
#Import the dataset

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("Employee_dataset.csv")
print(data.head())
```

```python
# Univariant EDA
```

```python
# Analyze Age Distribution of Employees (Histogram)

sns.histplot(data['age'])
```

```python
# Analyze Gender Distribution of Employees (Count Plot)


sns.countplot(x ='gender_full', data = data, palette={'Male': 'skyblue', 'Female': 'lightpink'})
plt.title("Gender Distribution")
plt.show()
```

```python
# Identify Employee Status Percentage by Year (Pie Chart)

x = data['STATUS_YEAR'].value_counts()
plt.pie(x.values ,labels = x.index, autopct='%1.1f%%')
plt.show()
```

```python
# Bivariate EDA
```

```python
# Compare Length of Service Across Departments (Bar Plot)

plt.figure(figsize=(15,5))
sns.barplot(x='department_name', y='length_of_service', data=data, palette='Set3')
plt.xticks(rotation = 90)
plt.title("Length of Service Across Departments")
plt.show()
```

```python
# Relationship Between Age and Length of Service (Scatter Plot)

sns.scatterplot(x='length_of_service', y='age', data = data)
plt.show()
```

```python
# Employee Status Trend Over Years (Count Plot with Hue)

sns.countplot(x = 'STATUS_YEAR', hue='STATUS', data = data)
```

```python
# Multivariate Analysis
```

```python
# Relationship Between Age and Length of Service (Scatter Plot)

from sklearn import datasets, decomposition

iris = datasets.load_iris()
X = iris.data
y = iris.target

pca = decomposition.PCA(n_components=2)
X = pca.fit_transform(X)

sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y)
```

```python
# Correlation Between Numeric Features (Heatmap)

numeric_data = data.select_dtypes(include='number')
sns.heatmap(numeric_data.corr(),annot=True)
```
